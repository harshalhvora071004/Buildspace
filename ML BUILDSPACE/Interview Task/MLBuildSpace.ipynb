```python
#Task2.2 - Create a dataframe of the average tempo for each of the top 10 most occurring artists.
fo=df['artist'].value_counts()[:10].index
f1=df.groupby('artist')['tempo'].mean().to_dict()
f2 = {key: f1[key] for key in fo }
task1_2_2 = pd.DataFrame.from_dict(f2,orient="index",columns=["Mean Tempo"])
task1_2_2
```
