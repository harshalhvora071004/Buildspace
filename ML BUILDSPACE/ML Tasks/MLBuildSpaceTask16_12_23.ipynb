```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter Plot - Total Rooms vs. Total Bedrooms with gradient and color palette
sns.scatterplot(x='total_rooms', y='total_bedrooms', data=df, hue='total_bedrooms', palette='viridis', size='total_bedrooms')
plt.title('Total Rooms vs Total Bedrooms')
plt.xlabel('Total Rooms')
plt.ylabel('Total Bedrooms')
plt.show()
```
